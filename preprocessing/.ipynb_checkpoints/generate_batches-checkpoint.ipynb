```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
N = 100
n_trials = 500

mu_true = 100
```

```python
X = np.empty((n_trials, N))

for i in range(n_trials):
    for n in range(N):
        x = np.random.poisson(lam=mu_true, size=n+1)
        X[i, n] = np.abs(np.mean(x)-mu_true)
        
n = np.array(range(N))+1
x = np.mean(X, axis=0)
```

```python
def func(x, a):
    return a / np.sqrt(x)

popt, pcov = curve_fit(func, n, x)
```

```python
plt.figure()
plt.plot(np.mean(X, axis=0))
plt.plot(n, func(n, *popt))
plt.show()
```
